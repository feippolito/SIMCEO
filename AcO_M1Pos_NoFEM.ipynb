```python
import dos
```

```python
path = './dos/AcO_M1Pos_noFEM'
```

```python
# ini
sim = dos.DOS(path, 10, show_timing = 2)
```

```python
#run, wait, terminate
sim.start()
sim.join()
sim.terminate()
```

# plot logs
---

```python
import matplotlib.pyplot as plt
try:
    from jupyterthemes import jtplot
    jtplot.style()
except:
    pass

import numpy as np
```

```python
sim.logs
```

```python
RTxyz = sim.logs.entries['AcO_controller']['M1 RTxyz'].timeSeries[1]
M1Pos_out = sim.logs.entries['M1PosHP0']['M1PosHP0_out'].timeSeries[1]
```

```python
reorder = {
    'segments': 7,
      'groups': 3,
      'inputs': 2}
```

```python
RTxyz_reorder = np.empty(RTxyz.shape)
for k in range(RTxyz.shape[0]):
    RTxyz_reorder[k,:] = np.hstack([x[i].reshape(1,-1) for i in range(reorder['segments'])
                    for x in np.split(RTxyz[k].reshape(-1,reorder['groups']),
                                      reorder['inputs'])])
    
RTxyz_reorder = RTxyz_reorder.reshape(50,7,6)
M1Pos_out_reorder = M1Pos_out.reshape(50,7,6)
```

```python
# Select
seg =  1    #1 - 7
rbm = 'Rx'  # Rx Ry Rz Tx Ty Tz

# Plot ------
table = ['Tx','Ty','Tz','Rx','Ry','Rz']
rbm_index = table.index(rbm)

plt.figure(figsize=(15,10))
ax1 = plt.subplot(2,1,1)
ax1.set_title('AcO output - ' + rbm)
plt.plot(RTxyz_reorder[:,seg - 1, rbm_index])
ax2 = plt.subplot(2,1,2)
ax2.set_title('HP output - ' + rbm)
plt.plot(M1Pos_out_reorder[:,seg - 1, rbm_index])
plt.show()
```
